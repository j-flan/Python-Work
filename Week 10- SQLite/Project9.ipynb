```python
import sqlite3
#access database
conn = sqlite3.connect('Practice.db')
cur = conn.cursor()
```

<h4>Ex 1:</h4>This will return all records, showing only the two columns CategoryName and Description.

```python
cur.execute('SELECT CategoryName, Description FROM Categories') #query
for col in cur.description:
    print (col[0], end = "  ") #print column headers
print()

for row in cur: #print rows
    print(row) 
```

<h4>Ex 2:</h4>This will return all records, showing only the FirstName, LastName, and HireDate of all employees (in the Employees table) with the Title of Sales Representative

```python
cur.execute('SELECT FirstName, LastName, HireDate FROM Employees WHERE Title = "Sales Representative"')
for col in cur.description:
    print (col[0], end = "  ")
print()

for row in cur:
    print(row)
```

<h4>Ex 3:</h4> This uses the same columns as Ex. 2, but allows the user to select the Title. Available titles are: Sales Representative, Sales Manager, Vice President, Inside Sales Coordinator.

```python
tag = input("Enter title to search from: Sales Representative, Sales Manager, Vice President, Inside Sales Coordinator.")
cur.execute('SELECT FirstName, LastName, HireDate FROM Employees WHERE Title = ?', (tag,))
for col in cur.description:
    print (col[0], end = "  ")
print()

for row in cur:
    print(row)
```

<h4>Ex 4:</h4> This uses the same columns as Ex. 2, but allows the user to select the Title and the Country. Available countries are: USA, UK

```python
tag = input("Enter title to search from: Sales Representative, Sales Manager, Vice President,Inside Sales Coordinator.")
place = input("Enter Employee Country: USA or UK")
cur.execute('SELECT FirstName, LastName, HireDate FROM Employees WHERE Title = ? AND Country = ?', (tag, place,))
for col in cur.description:
    print (col[0], end = "  ")
print()

for row in cur:
    print(row)
```

<h4>Ex 5:</h4>This displays all orders placed by a specific employee. The EmployeeID for this employee is 5 (Steven Buchanan), and displays only the OrderID and the OrderData.

```python
cur.execute('SELECT OrderID, OrderDate, ShippedDate, ShipName, ShipAddress,\
        ShipCity,ShipPostalCode, ShipCountry FROM Orders WHERE EmployeeID = "5"')
for col in cur.description:
    print (col[0], end = "  ")
print()

for row in cur:
    print(row)
```

<h4>Ex 6.</h4>This displays SupplierID, ContactName, and ContactTitle for those Suppliers whose ContactTitle is not Marketing Manager.

```python
cur.execute('SELECT ProductID, ProductName FROM Products\
        JOIN Suppliers ON Products.SupplierID = Suppliers.SupplierID\
        WHERE ContactTitle NOT LIKE "%Marketing Manager%"')
for col in cur.description:
    print (col[0], end = "  ")
print()

for row in cur:
    print(row)
```

<h4>Ex 7.</h4>This displays ProductID and ProductName for those Products where ProductName includes the string “queso".

```python
cur.execute('SELECT ProductID, ProductName FROM Products\
            WHERE ProductName LIKE "%queso%"')
for col in cur.description:
    print (col[0], end = "  ")
print()

for row in cur:
    print(row)
```

<h4>Ex 8.</h4>This displays OrderID, CustomerID, and ShipCountry for Orders where the ShipCountry is either France or Belgium.

```python
cur.execute('SELECT OrderID, CustomerID, ShipCountry FROM Orders\
            WHERE ShipCountry ="France" OR ShipCountry = "Belgium"\
            ORDER BY ShipCountry')
for col in cur.description:
    print (col[0], end = "  ")
print()

for row in cur:
    print(row)
```

<h4>Ex 9.</h4>This displays the FirstName, LastName, Title, and BirthDate for Employees, listing the oldest employees first.

```python
cur.execute('SELECT FirstName, LastName, Title, BirthDate FROM Employees\
            ORDER BY BirthDate')
for col in cur.description:
    print (col[0], end = "  ")
print()

for row in cur:
    print(row)
```

<h4>Ex 10.</h4>This displays the OrderID, ProductID, UnitPrice, Quantity, and a new field called TotalPrice for OrderDetails, that multiplies these two together.

```python
cur.execute('''SELECT OrderID, ProductID, UnitPrice, Quantity, UnitPrice * Quantity AS TotalPrice\
            FROM OrderDetails''')

for col in cur.description:
    print (col[0], end = "  ")
print()

for row in cur:
    print(row)
```

<h4>Ex 11.</h4>This displays the total number of customers in the Customers table.

```python
cur.execute('''SELECT COUNT(CustomerID) AS customers FROM Customers''')

for col in cur.description:
    print (col[0], end = "  ")
print()

for row in cur:
    print(row)
```

<h4>Ex 12.</h4>This displays an alphabetical list of (unique) countries from the Customers table.

```python
cur.execute('SELECT DISTINCT Country FROM Customers ORDER BY Country')

for col in cur.description:
    print (col[0], end = "  ")
print()

for row in cur:
    print(row)
```

<h4>Ex 13.</h4>This displays a list of ContactTitle and a count for each ContactTitle and orders the list in descending order by count.

```python
cur.execute('SELECT ContactTitle, COUNT(ContactTitle) AS Count\
            FROM Suppliers\
            GROUP BY ContactTitle\
            ORDER BY Count DESC')

for col in cur.description:
    print (col[0], end = "  ")
print()

for row in cur:
    print(row)
```
