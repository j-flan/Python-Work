```python
import sqlite3
#access database
conn = sqlite3.connect('Practice.db')
cur = conn.cursor()
```

<br><strong>Ex 1.</strong> This program shows for each Product, the associated Supplier, then sorts the result by ProductID

```python
cur.execute('SELECT ProductID, ProductName, CompanyName FROM Products\
            JOIN Suppliers ON Products.SupplierID = Suppliers.SupplierID\
            ORDER BY ProductID') #query
for col in cur.description:
    print (col[0], end = "  ") #print column headers
print()

for row in cur: #print rows
    print(row) 
```

<br><strong>Ex 2.</strong> This program that displays a list of the Orders that were shipped by the Shipper: Federal Shipping, and the date they were ordered.

```python
cur.execute('SELECT OrderID, OrderDate, CompanyName FROM Orders\
            JOIN Shippers ON Orders.ShipVia = Shippers.ShipperID\
           WHERE CompanyName = "Federal Shipping"') #query
for col in cur.description:
    print (col[0], end = "  ") #print column headers
print()

for row in cur: #print rows
    print(row) 
```

<br><strong>Ex 3.</strong> This program displays the total number of products in each category, and sorts the number of products descending

```python
cur.execute('SELECT COUNT(ProductName) AS Amount, CategoryName FROM Products\
            JOIN Categories ON Products.CategoryID = Categories.CategoryID\
            GROUP BY CategoryName\
            ORDER BY Amount DESC') #query
for col in cur.description:
    print (col[0], end = "  ") #print column headers
print()

for row in cur: #print rows
    print(row) 
```

<br><strong>Ex 4.</strong> This program displays the total number of customers in the Customers table per Country and City.

```python
cur.execute('SELECT COUNT(CustomerID) AS Customers, Country, City FROM Customers\
            GROUP BY Country, City') #query
for col in cur.description:
    print (col[0], end = "  ") #print column headers
print()

for row in cur: #print rows
    print(row) 
```

<br><strong>Ex 5.</strong> This program displays the Products in inventory that should be reordered, and displays the ProductID, ProductName, UnitsInStock, and ReorderLevel. It ignores the products already on order, unless the order amount is still less than the reorder level.

```python
cur.execute('SELECT ProductID, ProductName, UnitsInStock, ReorderLevel FROM Products\
            WHERE UnitsInStock < ReorderLevel AND (UnitsOnOrder + UnitsInStock) <= ReorderLevel\
            ORDER BY ProductName') #query
for col in cur.description:
    print (col[0], end = "  ") #print column headers
print()

for row in cur: #print rows
    print(row) 
```

<br><strong>Ex 6.</strong>This program is the same as above,but adds a check for producs that are discontinued.

```python
cur.execute('SELECT ProductID, ProductName, UnitsInStock, ReorderLevel FROM Products\
            WHERE UnitsInStock < ReorderLevel AND (UnitsOnOrder + UnitsInStock) <= ReorderLevel\
            AND Discontinued = 0\
            ORDER BY ProductName') #query
for col in cur.description:
    print (col[0], end = "  ") #print column headers
print()

for row in cur: #print rows
    print(row) 
```

<br><strong>Ex 7.</strong> This program displays a list of all customers sorted alphabetically by region, sorted by region, then CustomerID. Customers with no region are not listed.

<br><strong>Ex 8.</strong> This program displays the three highest values for average overall freight for ship countries, in descending order, by average freight.

```python
cur.execute('SELECT ROUND((Freight / Quantity),2) AS "AVG-Freight", ShipCountry FROM Orders\
            JOIN OrderDetails ON Orders.OrderID = OrderDetails.OrderID\
            GROUP BY ShipCountry\
            ORDER BY "AVG-Freight" DESC\
            LIMIT 3') #query
for col in cur.description:
    print (col[0], end = "  ") #print column headers
print()

for row in cur: #print rows
    print(row) 
```
