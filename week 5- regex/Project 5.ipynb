<h4>1. This program sorts romeo-full.txt by word length in decending order and secondly by descending lexographic order.Then, prints the top 6 words.</h4>

```python
import string
try:
    fname = open ('romeo-full.txt')
except:
    print('File not found')
    exit()
#get words line-by-line
t = list()
for line in fname:
    line = line.replace('-', ' ')
    line = line.translate(line.maketrans('','', string.punctuation))
    words = line.split()
    #make list of words with keys and sort by key, then word
    for word in words:
        t.append((len(word), word))
        t.sort(reverse = True)         
print(t[:6])
```

<br><h4>2. This program finds the last letter of each word in the text file, then counts and orders them from most to least popular, and outputs the top 6.</h4>

```python
#the most popular ending letter in romeo-full.txt
import re
import string
try:
    fname = open ('romeo-full.txt')
except:
    print('File not found')
    exit()
#get ending letters line-by-line
t = dict()
for line in fname:
    #replace hyphens for spaces
    line = line.replace('-', ' ')
    #remove punctuation
    line = line.translate(line.maketrans('','', string.punctuation))
    #find last letters of words on line
    letter = re.findall('[a-z] ', line)
    #make dict of letters
    for i in letter:
        if i not in t:
            t[i] = 1
        else:
            t[i] += 1
#make list of letters and sort descending
order = list(t.items())
lzt = list()
for key ,val in order:
    lzt.append((val,key))
lzt.sort(reverse = True)         
print(lzt[:6])
    
```

<br><h4>3. This program uses a regular expression pattern to match all words ending in s, and returns the list of (unique) words from wordlist.txt</h4>

```python
import re
import string
try:
    fname = open ('wordlist.txt')
except:
    print('File not found')
    exit()

t = []
for word in fname:
    word = word.strip()
    #find words that end in 's'
    x = re.findall('([A-z]+s)', word)
    #add them to the list if they  have content and are unique
    if len(x) > 0 and x not in t:
        t.append(x)
print(t)
```

<br><h4>5.This program uses a regular expression pattern to match all words with ss anywhere in the word,  and returns the list of (unique) words using wordlist.txt</h4>

```python
import re
import string
try:
    fname = open ('wordlist.txt')
except:
    print('File not found')
    exit()

t = []
for word in fname:
    word = word.strip()
    #find words that contain 'ss'
    x = re.findall('([A-z]*ss[A-z]*)', word)
    #add them to the list if they  have content and are unique
    if len(x) > 0 and x not in t:
        t.append(x)
print(t)
```

<br><h4>6.This program uses a regular expression pattern to match all words beginning and ending with the letter a, and returns the list of (unique) words using wordlist.txt.</h4> 

```python
import re
import string
try:
    fname = open ('wordlist.txt')
except:
    print('File not found')
    exit()

t = []
for word in fname:
    word = word.strip()
    #find words that start and end in 'a'
    x = re.findall('(a[A-z]*a)', word)
    #add them to the list if they  have content and are unique
    if len(x) > 0 and x not in t:
        t.append(x)
print(t)
```

<br><h4>7.This program uses a function that extracts the domain name from a URL. The domain name is the part of the URL that comes after http://   but before the next /  . (ie techtutoronline.com, using httplist.txt</h4>

```python
import re
import string
try:
    fname = open ('httplist.txt')
except:
    print('File not found')
    exit()

t = []
for word in fname:
    word = word.strip()    
    #find words that start and end in 'a'
    x = re.findall('://([A-Za-z.]+[comrgedu])/', word)
    #add them to the list if they  have content and are unique
    if len(x) > 0 and x not in t:
        t.append(x)
print(t)
```
